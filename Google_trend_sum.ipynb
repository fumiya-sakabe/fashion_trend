```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Fashion_Trend/Botomus_trend.csv')

df.head()
```

```python
#step1 月の文字列データを月の日時データとして変換する
#step2 月の日時データを春夏秋冬に変更する関数をつくる
#step3 applyメソッドでstep2で作った関数を適応する
```

```python
df['月'] = pd.to_datetime(df['月']) #このままだと'月'が文字列になってしまっているため、to_datetimeを使って日時データに変換している
```

```python
def month_to_season(month):    #月と季節を結びつけるために月の日時データを春夏秋冬に変更する関数を作る
    if month == 1:               #monthが1の時
        return 'winter'            #winterを返す
    elif month == 2:             #monthが2の時
        return 'winter'            #winterを返す
    elif month == 3:             #monthが3の時
        return 'spring'            #springを返す
    elif month == 4:             #monthが4の時
        return 'spring'            #springを返す
    elif month == 5:             #monthが5の時
        return 'spring'            #springを返す
    elif month == 6:             #monthが6の時
        return 'summer'          #summerを返す
    elif month == 7:             #monthが7の時
        return 'summer'          #summerを返す
    elif month == 8:             #monthが8の時
        return 'summer'          #summerを返す
    elif month == 9:             #monthが9の時
        return 'autumn'          #autumnを返す
    elif month == 10:            #monthが10の時
        return 'autumn'           #autumnを返す
    elif month == 11:            #monthが11の時
        return 'autumn'           #autumnを返す
    else:                         #それ以外の場合
        return 'winter'           #winterを返す
```

```python
def month_to_season_refactoring(month):
    if month == 1 or month == 2 or month == 12:
        return 'winter'
    elif month == 3 or month == 4 or month == 5:
        return 'spring'
    elif month == 6 or month == 7 or month or 8:
        return 'summer'
    else:
        return 'autumn'
```

```python
def month_to_season_refactoring2(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:  # December, January, February
        return 'Winter'
```

```python
df['only_month'] = df['月'].dt.month                      #'月'から年を取り除き、月だけを抽出
df['season'] = df['only_month'].apply(month_to_season)  #applyメソッドを使って1行ごとにmonth_to_seasonを呼び出して'season'に格納
```

```python
#step4 アルファベット順で処理されてしまうため、文字列になっているseasonを'冬''春''夏''秋'で処理順で設定する
#step5 各季節ごとに平均値を出す
#step6 前のシーズンのデータと比べてどれだけ変化しているかを調べる
#step7 その時の変化が一番高いところをトレンドとする
```

```python
df['season'] = pd.Categorical(df['season'], categories=['winter', 'spring', 'summer', 'autumn'], ordered=True)
```

```python
# 'Year'列を追加
df['Year'] = df['月'].dt.year
```

```python
df_pivot = df.pivot_table(index=['Year', 'season'], values=df.columns[1:6], aggfunc='mean')
```

```python
growth_rate = df_pivot.pct_change(1).dropna()
```

```python
growth_rate
```

```python
trending_items = growth_rate.idxmax(axis=1).reset_index()
trending_items.columns = ['Year', 'season', 'Trending Item']
```
