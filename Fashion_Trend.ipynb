```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
df = pd.read_csv("fashion_trend_calu.csv")
```

```python
df
```

```python
df.replace({"Botomus": {"スウェットパンツ: (日本)": 0, "デニムパンツ: (日本)": 1, "カーゴパンツ: (日本)": 2,"スラックス: (日本)": 3, "スキニーパンツ: (日本)": 4}}, inplace=True)
df.replace({"Coat": {"Pコート": 5, "チェスターコート": 6, "トレンチコート": 7}}, inplace=True)
df.replace({"Shirt": {"Tシャツ": 8, "タンクトップ": 9}}, inplace=True)
df.replace({"Botomus(半年前)": {"スウェットパンツ: (日本)": 0, "デニムパンツ: (日本)": 1, "カーゴパンツ: (日本)": 2,"スラックス: (日本)": 3, "スキニーパンツ: (日本)": 4}}, inplace=True)
df.replace({"Coat(半年前)": {"Pコート": 5, "チェスターコート": 6, "トレンチコート": 7}}, inplace=True)
df.replace({"Shirt(半年前)": {"Tシャツ": 8, "タンクトップ": 9}}, inplace=True)
```

```python
a = df
```

```python
b = a.fillna(0)
```

```python
b
```

```python
x = b[['Coat', 'Shirt', 'Botomus(半年前)', 'Coat(半年前)', 'Shirt(半年前)']]
y = b[['Botomus']]

s = b[['Botomus', 'Shirt', 'Botomus(半年前)', 'Coat(半年前)', 'Shirt(半年前)']]
t = b[['Coat']]

m = b[['Botomus', 'Coat', 'Botomus(半年前)', 'Coat(半年前)', 'Shirt(半年前)']]
n = b[['Shirt']]
```

```python
x
```

```python
y
```

```python
s
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

s_train, s_test, t_train, t_test = train_test_split(s, t, test_size = 0.2)

m_train, m_test, n_train, n_test = train_test_split(m, n, test_size = 0.2)
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC()

svm.fit(x_train, y_train)
```

```python
test = svm.predict(x_test)
```

```python
score = accuracy_score(y_test, test)
print(f"正解率：{score * 100}%")
```

```python
model = RandomForestClassifier()

model.fit(x_train, y_train)
```

```python
test = model.predict(x_test)
```

```python
score = accuracy_score(y_test, test)
print(f"正解率：{score * 100}%")
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC()

svm.fit(s_train, t_train)
```

```python
test = svm.predict(s_test)
```

```python
score = accuracy_score(y_test, test)
print(f"正解率：{score * 100}%")
```

```python
model = RandomForestClassifier()

model.fit(s_train, t_train)
```

```python
test = model.predict(s_test)
```

```python
score = accuracy_score(t_test, test)
print(f"正解率：{score * 100}%")
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC()

svm.fit(m_train, n_train)
```

```python
test = svm.predict(m_test)
```

```python
score = accuracy_score(n_test, test)
print(f"正解率：{score * 100}%")
```

```python
test = model.predict(m_test)
```
